```python
import matplotlib.pyplot as plt
import cv2
import random
import os

def display_images(category, num_samples=3):
    category_path = os.path.join(dataset_path, category)
    images = os.listdir(category_path)
    sample_images = random.sample(images, num_samples)

    plt.figure(figsize=(12, 4))
    for i, img_name in enumerate(sample_images):
        img_path = os.path.join(category_path, img_name)
        img = cv2.imread(img_path)

        # Check if image was loaded successfully
        if img is not None:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
            plt.subplot(1, num_samples, i + 1)
            plt.imshow(img)
            plt.title(category)
            plt.axis("off")
        else:
            print(f"Could not read image: {img_path}")

    plt.show()

# Display samples from each category
for cat in categories:
    display_images(cat)
```

```python
import cv2
import numpy as np

IMG_SIZE = (256, 256)  # Adjust as needed

def preprocess_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    img = cv2.resize(img, IMG_SIZE)  # Resize to fixed dimensions
    img = img / 255.0  # Normalize to [0,1]
    return img
```

```python
import os

blurred_path = os.path.join(dataset_path, "motion_blurred")  # Change to 'defocused_blurred' if needed
sharp_path = os.path.join(dataset_path, "sharp")

blurred_images = sorted(os.listdir(blurred_path))
sharp_images = sorted(os.listdir(sharp_path))

X = []  # Inputs (Blurred)
Y = []  # Outputs (Sharp)

for b_img, s_img in zip(blurred_images, sharp_images):
    blurred_img_path = os.path.join(blurred_path, b_img)
    sharp_img_path = os.path.join(sharp_path, s_img)

    X.append(preprocess_image(blurred_img_path))
    Y.append(preprocess_image(sharp_img_path))

X = np.array(X)
Y = np.array(Y)

print("Dataset shape:", X.shape, Y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Training set shape:", X_train.shape, Y_train.shape)
print("Testing set shape:", X_test.shape, Y_test.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, BatchNormalization, Activation, Input, UpSampling2D
from tensorflow.keras.models import Model

def build_sharpening_cnn(input_shape=(256, 256, 3)):
    inputs = Input(shape=input_shape)

    # Encoder
    x = Conv2D(64, (3, 3), padding="same")(inputs)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    x = Conv2D(128, (3, 3), padding="same")(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    # Decoder
    x = UpSampling2D((2, 2))(x)
    x = Conv2D(64, (3, 3), padding="same")(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    x = Conv2D(3, (3, 3), padding="same", activation="sigmoid")(x)  # Output: sharpened image

    model = Model(inputs, x)
    return model

# Initialize model
model = build_sharpening_cnn()
model.summary()
```

```python
model.compile(optimizer="adam", loss="mse", metrics=["accuracy"])

# Train the model
history = model.fit(X_train, Y_train, epochs=20, batch_size=16, validation_data=(X_test, Y_test))
```
